```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
df = pd.read_csv("../Aras/house_a_combined_dataset.csv")
```

```python
df.drop(columns=['Unnamed: 0',"Hour"],inplace=True)
df
```

```python
sensor_data = df.values  

model = IsolationForest(contamination=0.01)
anomalies = model.fit_predict(sensor_data)

df['anomaly'] = anomalies
```
