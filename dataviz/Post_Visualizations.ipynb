Threshold Values for House A and House B

```python
import matplotlib.pyplot as plt
import numpy as np

reconstruction_errors = np.load("../Pipeline/House A files/reconstruction_errors_ha.npy")

threshold = np.percentile(reconstruction_errors, 95)

filtered_reconstruction_errors = reconstruction_errors[reconstruction_errors < 0.001]

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

axes[0].hist(reconstruction_errors, bins=300, color='blue', edgecolor='black')
axes[0].axvline(x=threshold, color='red', linestyle='--', linewidth=2, label=f'Threshold = {threshold:.6f}')
axes[0].set_title('Histogram of Reconstruction Errors (Full) - House A')
axes[0].set_xlabel('Reconstruction Error')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

axes[1].hist(filtered_reconstruction_errors, bins=300, color='blue', edgecolor='black')
axes[1].axvline(x=threshold, color='red', linestyle='--', linewidth=2, label=f'Threshold = {threshold:.6f}')
axes[1].set_title('Histogram of Reconstruction Errors (Zoomed In Vesrion)- House A')
axes[1].set_xlabel('Reconstruction Error')
axes[1].set_ylabel('Frequency')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt
import numpy as np

reconstruction_errors = np.load("../Pipeline/House B files/reconstruction_errors_hb.npy")

threshold = np.percentile(reconstruction_errors, 95)

filtered_reconstruction_errors = reconstruction_errors[reconstruction_errors < 0.001]

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

axes[0].hist(reconstruction_errors, bins=300, color='blue', edgecolor='black')
axes[0].axvline(x=threshold, color='red', linestyle='--', linewidth=2, label=f'Threshold = {threshold:.6f}')
axes[0].set_title('Histogram of Reconstruction Errors (Full) - House B')
axes[0].set_xlabel('Reconstruction Error')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

axes[1].hist(filtered_reconstruction_errors, bins=300, color='blue', edgecolor='black')
axes[1].axvline(x=threshold, color='red', linestyle='--', linewidth=2, label=f'Threshold = {threshold:.6f}')
axes[1].set_title('Histogram of Reconstruction Errors (Zoomed In Vesrion)- House B')
axes[1].set_xlabel('Reconstruction Error')
axes[1].set_ylabel('Frequency')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Time Series Analysis

```python
day_a = data_a.groupby('Hour')['Anomaly'].sum().reset_index()
day_b = data_b.groupby('Hour')['Anomaly'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=day_a, x='Hour', y='Anomaly', label='House A', marker='o')
sns.lineplot(data=day_b, x='Hour', y='Anomaly', label='House B', marker='o')
plt.title('Anomalies by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Anomaly Count')
plt.legend()
plt.show()
```

## Day Of The Week Analysis

```python
day_a = data_a.groupby('Day Of Week')['Anomaly'].sum().reset_index()
day_b = data_b.groupby('Day Of Week')['Anomaly'].sum().reset_index()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.barplot(data=day_a, x='Day Of Week', y='Anomaly', color='red', alpha=0.6)
plt.title('Total Anomalies by Day of Week (House A)')
plt.xlabel('Day Of Week')
plt.ylabel('Total Anomalies')

plt.subplot(1, 2, 2)
sns.barplot(data=day_b, x='Day Of Week', y='Anomaly', color='blue', alpha=0.6)
plt.title('Total Anomalies by Day of Week (House B)')
plt.xlabel('Day Of Week')
plt.ylabel('Total Anomalies')

plt.tight_layout()
plt.show()

```

##Sensor Analysis

```python
sensor_columns_a = [
    'photocell_wardrobe', 'photocell_couch', 'ir_tv_receiver',
    'force_couch_1', 'force_couch_2', 'distance_chair_1', 'distance_chair_2',
    'photocell_fridge', 'photocell_kitchen_drawer', 'photocell_wardrobe_2',
    'photocell_bathroom_cabinet', 'contact_house_door', 'contact_bathroom_door',
    'contact_shower_door', 'sonar_hall', 'sonar_kitchen', 'distance_tap',
    'distance_water_closet', 'temperature_kitchen', 'force_bed'
]

sensor_columns_b = [
    'contact_kitchen_cupboard_1', 'contact_kitchen_cupboard_2', 'contact_house_door',
    'contact_wardrobe_door_1', 'contact_wardrobe_door_2', 'contact_shower_door',
    'distance_tap', 'force_chair_1', 'force_chair_2', 'force_chair_3',
    'photocell_fridge', 'photocell_kitchen_drawer', 'pressure_mat_couch_1',
    'pressure_mat_couch_2', 'pressure_mat_bed_1', 'pressure_mat_bed_2',
    'pressure_mat_armchair', 'sonar_bathroom_door', 'sonar_kitchen', 'sonar_closet'
]

correlation_a = data_a[sensor_columns_a + ['Anomaly']].corr()['Anomaly'].sort_values()

correlation_b = data_b[sensor_columns_b + ['Anomaly']].corr()['Anomaly'].sort_values()
plt.figure(figsize=(12, 6))
correlation_a.drop('Anomaly').plot(kind='bar', color='blue', alpha=0.6)
plt.title('Correlation of Sensors with Anomalies (House A)')
plt.xlabel('Sensors')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
correlation_b.drop('Anomaly').plot(kind='bar', color='orange', alpha=0.6)
plt.title('Correlation of Sensors with Anomalies (House B)')
plt.xlabel('Sensors')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

House A top 100 sensors combination contributing to anomaly

```python
columns_to_exclude = ['Unnamed: 0', 'Resident1', 'Resident2', 'Hour', 'Week', 'Day Of Week']
df_filtered = data_a.drop(columns=columns_to_exclude)

anomalies = df_filtered[df_filtered['Anomaly'] == 1]

sensor_columns = anomalies.columns[:-1]  
anomalies['Sensor_Combination'] = anomalies[sensor_columns].apply(lambda row: tuple(row > 0), axis=1)

combination_counts = anomalies['Sensor_Combination'].value_counts()

top_100_combinations = combination_counts.head(100)

sensor_names = sensor_columns.tolist()  
combination_details = []

for combination, count in top_100_combinations.items():
    active_sensors = [sensor_names[i] for i, active in enumerate(combination) if active]
    combination_details.append({'Active_Sensors': active_sensors, 'Count': count})

top_100_combinations_df = pd.DataFrame(combination_details)

print("Top 100 Sensor Combinations Causing Maximum Anomalies:")
top_100_combinations_df
```

```python
from collections import Counter

all_active_sensors = [sensor for comb in top_100_combinations_df['Active_Sensors'] for sensor in comb]
sensor_counts = Counter(all_active_sensors)

plt.figure(figsize=(12, 12))
plt.pie(sensor_counts.values(), labels=sensor_counts.keys(), autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Sensor Contributions in Top 100 Anomalous Combinations')
plt.tight_layout()
plt.show()
```

```python
import networkx as nx

G = nx.Graph()
for comb in top_100_combinations_df['Active_Sensors']:
    G.add_edges_from([(comb[i], comb[j]) for i in range(len(comb)) for j in range(i + 1, len(comb))])

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, seed=42)  
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue')
nx.draw_networkx_edges(G, pos, alpha=0.5, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
plt.title('Network Graph of Sensors in Top 100 Anomalous Combinations')
plt.axis('off')
plt.show()
```
