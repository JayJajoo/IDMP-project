```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../Aras/House_B_Anomalies/house_b_detected_anomalies.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:,1:21]
```

```python
contamination = (dataset["Anomaly"]==True).sum()/len(dataset)
contamination
```

```python
isf = IsolationForest(contamination=contamination)
anomaly = isf.fit_predict(x)
```

```python
anomaly
```

```python
dataset["ISF anomaly"] = anomaly
```

```python
(dataset["ISF anomaly"]==-1).sum()
```
